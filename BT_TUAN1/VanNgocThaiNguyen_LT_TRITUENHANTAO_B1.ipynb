```python
#bt 13
a = list(map(int, input("Nhập mảng a: ").split()))
b = list(map(int, input("Nhập mảng b: ").split()))

min_len = min(len(a), len(b))
result = [a[i] + b[i] for i in range(min_len)]

if len(a) > len(b):
    result.extend(a[min_len:])
elif len(b) > len(a):
    result.extend(b[min_len:])

print("Mảng kết quả:", result)
```

```python
#b14
import numpy as np

#mảng a
m, n = 4, 5
a = np.random.randint(1, 100, (m, n))
print("Mảng a:\n", a)
#ptử dòng k
k = 1
print(f"Dòng {k}:", a[k])
#ptử cột k
k = 2
print(f"Cột {k}:", a[:, k])
#dòng có tổng lớn nhất
tong = [sum(a[i]) for i in range(m)]
print("Dòng có tổng lớn nhất:", np.argmax(tong))
#cột có tích nhỏ nhất
tich = [np.prod(a[:, j]) for j in range(n)]
print("Cột có tích nhỏ nhất:", np.argmin(tich))
#Phần tử dòng chẵn, cột lẻ
print("Dòng chẵn, cột lẻ:\n", a[::2, 1::2])
#TBC phần tử chẵn thuộc dòng lẻ
chan_dong_le = a[1::2][a[1::2] % 2 == 0]
print("TB chẵn ở dòng lẻ:", chan_dong_le.mean())
#tbc phần tử thuộc biên
bien = np.concatenate([a[0], a[-1], a[1:-1, 0], a[1:-1, -1]])
print("TB biên:", bien.mean())
#tbt tích phần tử không thuộc biên
khong_bien = a[1:-1, 1:-1]
print("TB tích không biên:", khong_bien.mean())
```

```python
#15
import numpy as np

sv_dtype = np.dtype([
    ('masv', 'U10'),
    ('ten', 'U20'),
    ('namsinh', int),
    ('dtb', float)
])

ds = np.array([
    ('02DH0001', 'Nguyen Van An', 2004, 7.5),
    ('02DH0002', 'Phan Thi Lan',  2003, 4.8),
    ('03CD0003', 'Tran Van Binh', 2004, 6.0),
    ('02DH0004', 'Phan Van Nam',  2002, 5.5),
    ('04TC0005', 'Le Thi Lan',    2005, 3.9)
], dtype=sv_dtype)

# c
print("len lop:", np.sum(ds['dtb'] >= 5))

# d
print("du 20 tuoi:")
for sv in ds[2026 - ds['namsinh'] >= 20]:
    print(sv)

# e
print("he DH:", np.sum([sv['masv'][2:4] == 'DH' for sv in ds]))

# f
print("ten Lan:", np.sum([sv['ten'].split()[-1] == 'Lan' for sv in ds]))

# g
print("ho Phan:", np.sum([sv['ten'].split()[0] == 'Phan' for sv in ds]))
```
