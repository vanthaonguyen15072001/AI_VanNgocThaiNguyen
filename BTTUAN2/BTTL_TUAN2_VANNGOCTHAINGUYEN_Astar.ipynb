```python
#b.	Thuật toán tìm kiếm A* trên đồ thị:
from collections import deque

class Graph:
    def __init__(self, adjac_lis):
        self.adjac_lis = adjac_lis

    def get_neighbors(self, v):
        return self.adjac_lis[v]

    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return H[n]

    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        closed_lst = set([])

        poo = {}
        poo[start] = 0

        par = {}
        par[start] = start

        while len(open_lst) > 0:
            n = None

            for v in open_lst:
                if n == None or poo[v] + self.h(v) < poo[n] + self.h(n):
                    n = v

            if n == None:
                print('Path does not exist!')
                return None

            if n == stop:
                reconst_path = []

                while par[n] != n:
                    reconst_path.append(n)
                    n = par[n]

                reconst_path.append(start)

                reconst_path.reverse()

                print('Path found: {}'.format(reconst_path))
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    poo[m] = poo[n] + weight

                else:
                    if poo[m] > poo[n] + weight:
                        poo[m] = poo[n] + weight
                        par[m] = n

                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        print('Path does not exist!')
        return None

adjac_lis = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('C', 1), ('D', 5)],
    'C': [('D', 12)]
}

graph1 = Graph(adjac_lis)
graph1.a_star_algorithm('A', 'C')
```

```python
#Bài 2. Cài đặt thuật giải A* tìm đường đi giữa 2 đỉnh bất kỳ trong đồ thị.
class DoThi:
    def __init__(self, ds_ke, h_dict):
        self.ds_ke = ds_ke
        self.h_dict = h_dict

    def lay_ln(self, v):
        return self.ds_ke.get(v, [])

    def h(self, n):
        return self.h_dict.get(n, 1)

    def a_sao(self, bd, kt):
        mo = {bd}
        dong = set()

        g = {bd: 0}
        cha = {bd: bd}

        while len(mo) > 0:
            n = None

            for v in mo:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n is None:
                print('Khong tim thay duong di!')
                return None

            if n == kt:
                duong_di = []

                while cha[n] != n:
                    duong_di.append(n)
                    n = cha[n]

                duong_di.append(bd)
                duong_di.reverse()

                print('Duong di: {}'.format(duong_di))
                return duong_di

            for (m, ts) in self.lay_ln(n):
                if m not in mo and m not in dong:
                    mo.add(m)
                    cha[m] = n
                    g[m] = g[n] + ts
                else:
                    if g[m] > g[n] + ts:
                        g[m] = g[n] + ts
                        cha[m] = n

                        if m in dong:
                            dong.remove(m)
                            mo.add(m)

            mo.remove(n)
            dong.add(n)

        print('Khong tim thay duong di!')
        return None

ds_ke = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('C', 1), ('D', 5)],
    'C': [('D', 12)],
    'D': []
}

h_dict = {
    'A': 1,
    'B': 1,
    'C': 1,
    'D': 1
}

dt = DoThi(ds_ke, h_dict)
dt.a_sao('A', 'D')
```

```python
from heapq import heappush, heappop

class Nut:
    def __init__(self, ht, ct, cp, h, dd):
        self.ht = ht
        self.ct = ct
        self.cp = cp
        self.h = h
        self.dd = dd

    def __lt__(self, khac):
        return (self.cp + self.h) < (khac.cp + khac.h)

def h_mst(ct, ht, bd, mt):
    if not ct:
        return mt[ht][bd]

    tap_ct = list(ct)
    min_ht = min(mt[ht][v] for v in tap_ct)
    min_bd = min(mt[v][bd] for v in tap_ct)

    if len(tap_ct) == 1:
        return min_ht + min_bd

    mst_cp = 0
    da_tham = [tap_ct[0]]
    chua_tham = tap_ct[1:]

    while chua_tham:
        canh_min = float('inf')
        u_chon = -1
        for u in da_tham:
            for v in chua_tham:
                if mt[u][v] < canh_min:
                    canh_min = mt[u][v]
                    u_chon = v
        mst_cp += canh_min
        da_tham.append(u_chon)
        chua_tham.remove(u_chon)

    return min_ht + mst_cp + min_bd

def a_sao_ngh(mt, bd):
    so_dinh = len(mt)
    ct_bd = tuple(i for i in range(so_dinh) if i != bd)

    h_bd = h_mst(ct_bd, bd, bd, mt)
    goc = Nut(bd, ct_bd, 0, h_bd, [bd])

    hd = []
    heappush(hd, goc)

    while hd:
        nut = heappop(hd)

        if not nut.ct:
            duong_cuoi = nut.dd + [bd]
            cp_cuoi = nut.cp + mt[nut.ht][bd]
            print(f"Duong di: {duong_cuoi}")
            print(f"Chi phi: {cp_cuoi}")
            return

        for v in nut.ct:
            ct_moi = tuple(x for x in nut.ct if x != v)
            cp_moi = nut.cp + mt[nut.ht][v]
            h_moi = h_mst(ct_moi, v, bd, mt)
            dd_moi = nut.dd + [v]

            con = Nut(v, ct_moi, cp_moi, h_moi, dd_moi)
            heappush(hd, con)

ma_tran_ts = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

bd = 0
a_sao_ngh(ma_tran_ts, bd)
```
