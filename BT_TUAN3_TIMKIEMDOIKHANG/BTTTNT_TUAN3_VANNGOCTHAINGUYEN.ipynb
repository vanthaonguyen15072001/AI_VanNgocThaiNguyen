```python
#Thuật giải Minimax ứng dụng vào bài toán TicTacToe
import copy
import math
import numpy

X = "X"
O = "O"
TRONG = None
nguoi = None
may = None

def tao_ban_co():
    return [[TRONG, TRONG, TRONG],
            [TRONG, TRONG, TRONG],
            [TRONG, TRONG, TRONG]]

def luot_cua_ai(ban_co):
    dem = 0
    for hang in ban_co:
        for o in hang:
            if o:
                dem += 1
    if dem % 2 != 0:
        return may
    return nguoi

def cac_buoc_di(ban_co):
    ds_buoc = set()
    kt = len(ban_co)
    for i in range(kt):
        for j in range(kt):
            if ban_co[i][j] == TRONG:
                ds_buoc.add((i, j))
    return ds_buoc

def tao_trang_thai_moi(ban_co, buoc_di):
    nguoi_dang_di = luot_cua_ai(ban_co)
    ban_co_moi = copy.deepcopy(ban_co)
    (i, j) = buoc_di
    ban_co_moi[i][j] = nguoi_dang_di
    return ban_co_moi

def kt_thang_ngang(ban_co):
    ng_thang = None
    kt = len(ban_co)
    for i in range(kt):
        ng_thang = ban_co[i][0]
        for j in range(kt):
            if ban_co[i][j] != ng_thang:
                ng_thang = None
        if ng_thang:
            return ng_thang
    return ng_thang

def kt_thang_doc(ban_co):
    ng_thang = None
    kt = len(ban_co)
    for i in range(kt):
        ng_thang = ban_co[0][i]
        for j in range(kt):
            if ban_co[j][i] != ng_thang:
                ng_thang = None
        if ng_thang:
            return ng_thang
    return ng_thang

def kt_thang_cheo(ban_co):
    ng_thang = None
    kt = len(ban_co)
    ng_thang = ban_co[0][0]
    for i in range(kt):
        if ban_co[i][i] != ng_thang:
            ng_thang = None
    if ng_thang:
        return ng_thang
    ng_thang = ban_co[0][kt - 1]
    for i in range(kt):
        j = kt - 1 - i
        if ban_co[i][j] != ng_thang:
            ng_thang = None
    return ng_thang

def tim_nguoi_thang(ban_co):
    ng_thang = kt_thang_ngang(ban_co) or kt_thang_doc(ban_co) or kt_thang_cheo(ban_co) or None
    return ng_thang

def kt_ket_thuc(ban_co):
    if tim_nguoi_thang(ban_co) != None:
        return True
    for hang in ban_co:
        for o in hang:
            if o == TRONG:
                return False
    return True

def tinh_diem(ban_co):
    ng_thang = tim_nguoi_thang(ban_co)
    if ng_thang == X:
        return 1
    elif ng_thang == O:
        return -1
    return 0

def gia_tri_max(tt):
    if kt_ket_thuc(tt):
        return tinh_diem(tt)
    diem_max = -math.inf
    for buoc_di in cac_buoc_di(tt):
        diem_max = max(diem_max, gia_tri_min(tao_trang_thai_moi(tt, buoc_di)))
    return diem_max

def gia_tri_min(tt):
    if kt_ket_thuc(tt):
        return tinh_diem(tt)
    diem_min = math.inf
    for buoc_di in cac_buoc_di(tt):
        diem_min = min(diem_min, gia_tri_max(tao_trang_thai_moi(tt, buoc_di)))
    return diem_min

def minimax(ban_co):
    nguoi_dang_di = luot_cua_ai(ban_co)
    buoc_chon = None
    if nguoi_dang_di == X:
        diem_max = -math.inf
        for buoc_di in cac_buoc_di(ban_co):
            diem = gia_tri_min(tao_trang_thai_moi(ban_co, buoc_di))
            if diem > diem_max:
                diem_max = diem
                buoc_chon = buoc_di
    else:
        diem_min = math.inf
        for buoc_di in cac_buoc_di(ban_co):
            diem = gia_tri_max(tao_trang_thai_moi(ban_co, buoc_di))
            if diem < diem_min:
                diem_min = diem
                buoc_chon = buoc_di
    return buoc_chon

if __name__ == "__main__":
    ban_co = tao_ban_co()
    print("Chon nguoi choi (X hoac O):")
    nguoi = input().upper()
    if nguoi == "X":
        may = "O"
    else:
        may = "X"

    while True:
        het_game = kt_ket_thuc(ban_co)
        nguoi_dang_di = luot_cua_ai(ban_co)

        if het_game:
            ng_thang = tim_nguoi_thang(ban_co)
            if ng_thang is None:
                print("Game Over: Hoa.")
            else:
                print(f"Game Over: {ng_thang} thang.")
            break
        else:
            if nguoi != nguoi_dang_di:
                buoc_di = minimax(ban_co)
                ban_co = tao_trang_thai_moi(ban_co, buoc_di)
                print(numpy.array(ban_co))
            else:
                print("Nhap toa do de di (Hang, Cot) tu 0 den 2:")
                while True:
                    try:
                        i = int(input("Hang: "))
                        j = int(input("Cot: "))
                        if 0 <= i < 3 and 0 <= j < 3:
                            break
                        else:
                            print("Toa do khong hop le. Vui long nhap so tu 0 den 2.")
                    except ValueError:
                        print("Dau vao khong hop le. Vui long nhap so nguyen.")

                if ban_co[i][j] == TRONG:
                    ban_co = tao_trang_thai_moi(ban_co, (i, j))
                    print(numpy.array(ban_co))
                else:
                    print("O nay da co nguoi di. Vui long chon o khac.")
```

```python
#b. Thuật toán cắt tỉa Alpha-beta với ứng dụng vào bài toán TicTacToe
import os, math

def KiemTraThang(ban_co):
    """
    Trả về người chiến thắng trên bàn cờ hiện tại nếu có, ngược lại trả về None.
    """
    #ngang
    if ban_co[0] == ban_co[1] and ban_co[1] == ban_co[2]:
        return ban_co[0]
    elif ban_co[3] == ban_co[4] and ban_co[4] == ban_co[5]:
        return ban_co[3]
    elif ban_co[6] == ban_co[7] and ban_co[7] == ban_co[8]:
        return ban_co[6]
    #dọc
    elif ban_co[0] == ban_co[3] and ban_co[3] == ban_co[6]:
        return ban_co[0]
    elif ban_co[1] == ban_co[4] and ban_co[4] == ban_co[7]:
        return ban_co[1]
    elif ban_co[2] == ban_co[5] and ban_co[5] == ban_co[8]:
        return ban_co[2]
    #chéo
    elif ban_co[0] == ban_co[4] and ban_co[4] == ban_co[8]:
        return ban_co[0]
    elif ban_co[2] == ban_co[4] and ban_co[4] == ban_co[6]:
        return ban_co[2]

    return None

def InBanCo(ban_co):
    """
    Xóa console và in bàn cờ hiện tại.
    """
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f'''
     {ban_co[0]} | {ban_co[1]} | {ban_co[2]}
    {ban_co[3]} | {ban_co[4]} | {ban_co[5]}
    {ban_co[6]} | {ban_co[7]} | {ban_co[8]}
    ''')

def LayOTrong(ban_co):
    """
    Trả về danh sách các ô còn trống trên bàn cờ.
    """
    o_trong = list()
    for o in ban_co:
        if o != "X" and o != "O":
            o_trong.append(o)
    return o_trong

def minimax(ban_co, do_sau, alpha, beta, la_max):
    """
    Thuật toán AI chịu trách nhiệm chọn nước đi tốt nhất với cắt tỉa Alpha-Beta.
    """
    nguoi_thang = KiemTraThang(ban_co)
    if nguoi_thang != None:
        return 10 - do_sau if nguoi_thang == "X" else -10 + do_sau

    if len(LayOTrong(ban_co)) == 0:
        return 0

    if la_max:
        gt_max = -math.inf
        for o in LayOTrong(ban_co):
            ban_co[o - 1] = "X"
            gt = minimax(ban_co, do_sau + 1, alpha, beta, False)
            gt_max = max(gt_max, gt)
            alpha = max(alpha, gt)
            ban_co[o - 1] = o

            if beta <= alpha:
                break
        return gt_max
    else:
        gt_min = math.inf
        for o in LayOTrong(ban_co):
            ban_co[o - 1] = "O"
            gt = minimax(ban_co, do_sau + 1, alpha, beta, True)
            gt_min = min(gt_min, gt)
            beta = min(beta, gt)
            ban_co[o - 1] = o

            if beta <= alpha:
                break
        return gt_min

def TimNuocDiTot(ban_co_ht, may):
    """
    Trả về nước đi tốt nhất có thể cho người chơi.
    """
    gt_tot_nhat = -math.inf if may == "X" else math.inf
    nuoc_di_tot = -1

    for o in LayOTrong(ban_co_ht):
        ban_co_ht[o - 1] = may
        gt_nuoc_di = minimax(ban_co_ht, 0, -math.inf, math.inf, False if may == "X" else True)
        ban_co_ht[o - 1] = o

        if (may == "X" and gt_nuoc_di > gt_tot_nhat):
            nuoc_di_tot = o
            gt_tot_nhat = gt_nuoc_di
        elif (may == "O" and gt_nuoc_di < gt_tot_nhat):
            nuoc_di_tot = o
            gt_tot_nhat = gt_nuoc_di

    return nuoc_di_tot

def main():
    nguoi_choi = input("Chơi X hay O? ").strip().upper()
    may = "O" if nguoi_choi == "X" else "X"
    tro_choi_ht = [*range(1, 10)]

    luot_ht = "X"
    dem = 0

    while True:
        if luot_ht == may:
            o_di = TimNuocDiTot(tro_choi_ht, may)
            tro_choi_ht[o_di - 1] = may
            luot_ht = nguoi_choi
        elif luot_ht == nguoi_choi:
            InBanCo(tro_choi_ht)
            while True:
                nguoi_nhap = int(input("Nhập ô muốn đánh (1-9): ").strip())
                if nguoi_nhap in tro_choi_ht:
                    tro_choi_ht[nguoi_nhap - 1] = nguoi_choi
                    luot_ht = may
                    break
                else:
                    InBanCo(tro_choi_ht)
                    print("Ô này không khả dụng hoặc đã được đánh.")

        if KiemTraThang(tro_choi_ht) != None:
            InBanCo(tro_choi_ht)
            print(f"{KiemTraThang(tro_choi_ht)} THẮNG!!!")
            break

        dem += 1
        if KiemTraThang(tro_choi_ht) == None and dem == 9:
            InBanCo(tro_choi_ht)
            print("Hòa.")
            break

if __name__ == "__main__":
    main()
```

```python
#BTVN BT3
import math

class TicTacToeNxN:
    def __init__(self, N, chuoi_de_thang, do_sau_toi_da=3):
        self.N = N
        self.chuoi_de_thang = chuoi_de_thang
        self.do_sau_toi_da = do_sau_toi_da
        self.ban_co = [[" " for _ in range(N)] for _ in range(N)]

    def in_ban_co(self):
        print("\n")
        for i in range(self.N):
            print(" | ".join(self.ban_co[i]))
            if i < self.N - 1:
                print("-" * (4 * self.N - 1))
        print("\n")

    def lay_o_trong(self):
        o_trong = []
        for i in range(self.N):
            for j in range(self.N):
                if self.ban_co[i][j] == " ":
                    o_trong.append((i, j))
        return o_trong

    def kiem_tra_thang(self, ban_co):
        huong_di = [(0, 1), (1, 0), (1, 1), (1, -1)]

        for r in range(self.N):
            for c in range(self.N):
                nguoi_choi = ban_co[r][c]
                if nguoi_choi == " ": continue

                for dr, dc in huong_di:
                    dem = 0
                    for k in range(self.chuoi_de_thang):
                        nr, nc = r + k * dr, c + k * dc
                        if 0 <= nr < self.N and 0 <= nc < self.N and ban_co[nr][nc] == nguoi_choi:
                            dem += 1
                        else:
                            break
                    if dem == self.chuoi_de_thang:
                        return nguoi_choi
        return None

    def danh_gia_trang_thai(self, ban_co):
        nguoi_thang = self.kiem_tra_thang(ban_co)
        if nguoi_thang == "X":
            return 1000
        elif nguoi_thang == "O":
            return -1000
        return 0

    def minimax(self, ban_co, do_sau, alpha, beta, la_max):
        diem_danh_gia = self.danh_gia_trang_thai(ban_co)
        o_trong = self.lay_o_trong()

        if diem_danh_gia == 1000: return diem_danh_gia - do_sau
        if diem_danh_gia == -1000: return diem_danh_gia + do_sau
        if len(o_trong) == 0: return 0
        if do_sau == self.do_sau_toi_da: return 0

        if la_max:
            gt_max = -math.inf
            for r, c in o_trong:
                ban_co[r][c] = "X"
                gt = self.minimax(ban_co, do_sau + 1, alpha, beta, False)
                ban_co[r][c] = " "

                gt_max = max(gt_max, gt)
                alpha = max(alpha, gt)
                if beta <= alpha: break
            return gt_max
        else:
            gt_min = math.inf
            for r, c in o_trong:
                ban_co[r][c] = "O"
                gt = self.minimax(ban_co, do_sau + 1, alpha, beta, True)
                ban_co[r][c] = " "

                gt_min = min(gt_min, gt)
                beta = min(beta, gt)
                if beta <= alpha: break
            return gt_min

    def tim_nuoc_di_tot_nhat(self, may):
        gt_tot_nhat = -math.inf if may == "X" else math.inf
        nuoc_di = None
        la_max_tiep_theo = False if may == "X" else True

        for r, c in self.lay_o_trong():
            self.ban_co[r][c] = may
            gt = self.minimax(self.ban_co, 0, -math.inf, math.inf, la_max_tiep_theo)
            self.ban_co[r][c] = " "

            if may == "X" and gt > gt_tot_nhat:
                gt_tot_nhat = gt
                nuoc_di = (r, c)
            elif may == "O" and gt < gt_tot_nhat:
                gt_tot_nhat = gt
                nuoc_di = (r, c)

        return nuoc_di

if __name__ == "__main__":
    N = int(input("Nhập kích thước bàn cờ N (ví dụ: 5, 10, 15): "))
    K = int(input("Nhập số lượng ô liên tiếp cần để thắng (ví dụ: 4, 5): "))

    if N <= 3:
        DO_SAU = 9
    elif N <= 5:
        DO_SAU = 4
    else:
        DO_SAU = 2

    game = TicTacToeNxN(N, K, DO_SAU)

    luot_cua_nguoi = "O"
    luot_cua_may = "X"
    luot_hien_tai = "X"

    print(f"BẮT ĐẦU TRÒ CHƠI {N}x{N}. Đạt {K} ô liên tiếp để thắng!")

    while True:
        game.in_ban_co()

        if luot_hien_tai == luot_cua_may:
            print("Máy đang suy nghĩ...")
            nuoc_di = game.tim_nuoc_di_tot_nhat(luot_cua_may)
            if nuoc_di:
                game.ban_co[nuoc_di[0]][nuoc_di[1]] = luot_cua_may
            luot_hien_tai = luot_cua_nguoi

        else:
            try:
                nhap = input(f"Đến lượt bạn ({luot_cua_nguoi}). Nhập Hàng Cột (cách nhau bởi dấu cách, từ 0 đến {N-1}): ")
                r, c = map(int, nhap.split())
                if 0 <= r < N and 0 <= c < N and game.ban_co[r][c] == " ":
                    game.ban_co[r][c] = luot_cua_nguoi
                    luot_hien_tai = luot_cua_may
                else:
                    print("Nước đi không hợp lệ, vui lòng nhập lại.")
            except ValueError:
                print("Lỗi định dạng. Vui lòng nhập số.")

        nguoi_thang = game.kiem_tra_thang(game.ban_co)
        if nguoi_thang:
            game.in_ban_co()
            print(f">>> {nguoi_thang} ĐÃ GIÀNH CHIẾN THẮNG! <<<")
            break
        elif len(game.lay_o_trong()) == 0:
            game.in_ban_co()
            print(">>> HÒA! Cả hai bên đều đã hết nước đi. <<<")
            break
```
